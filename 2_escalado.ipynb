# Escalado
Conjunto de datos: iris (incluido en sklearn)
Este conjunto de datos es compacto y perfecto para demostrar técnicas de escalado.

```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos
iris = load_iris(as_frame=True)
data = iris.data
```

```python
# Visualizar las estadísticas iniciales
print("Estadísticas iniciales:")
data.describe()
```

```python
# Aplicar MinMaxScaler
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=data.columns)
```

```python
# Aplicar StandardScaler
standard_scaler = StandardScaler()
data_standard = pd.DataFrame(standard_scaler.fit_transform(data), columns=data.columns)
```

### 1. ¿Qué diferencias encuentras entre las estadísticas originales y las del escalado MinMax?


Las estadísticas cambian porque el escalado MinMax transforma todos los valores al rango [0, 1], mientras que en los datos originales cada variable tiene su propio rango.

Los mínimos pasan a 0, los máximos a 1, y media y desviación estándar cambian de magnitud, aunque se conserva la posición relativa de los datos.

### 2. Crea una gráfica de la distribución de la variable 'sepal length (cm)' antes y después del escalado.

```python
# Cargar dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Escalado MinMax
scaler = MinMaxScaler()
data_minmax = pd.DataFrame(
    scaler.fit_transform(data),
    columns=data.columns
)

# Gráficas
plt.figure(figsize=(10,4))

# Antes del escalado
plt.subplot(1,2,1)
plt.hist(data["sepal length (cm)"], bins=30)
plt.title("Sepal length - Original")
plt.xlabel("sepal length (cm)")
plt.ylabel("Frecuencia")

# Después del escalado
plt.subplot(1,2,2)
plt.hist(data_minmax["sepal length (cm)"], bins=30)
plt.title("Sepal length - MinMax")
plt.xlabel("Valor escalado")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

### 3. Crea una gráfica de dispersión entre 'sepal length (cm)' y 'sepal width (cm)' con los datos escalados por StandardScaler.

```python
plt.figure(figsize=(6,5))

plt.scatter(
    data_standard["sepal length (cm)"],
    data_standard["sepal width (cm)"]
)

plt.title("Sepal length vs Sepal width (StandardScaler)")
plt.xlabel("sepal length (cm) - escalado")
plt.ylabel("sepal width (cm) - escalado")

plt.show()
```

### 4. ¿Qué técnica de escalado crees que es más adecuada para este conjunto de datos? ¿Por qué?

Para este conjunto de datos, StandardScaler es más adecuado porque las variables siguen distribuciones aproximadamente normales y no están acotadas naturalmente, por lo que centrar los datos en media 0 y desviación estándar 1 preserva mejor las relaciones entre variables.
Además, muchos algoritmos (regresión, SVM, k-means, PCA) funcionan mejor cuando los datos están estandarizados.
